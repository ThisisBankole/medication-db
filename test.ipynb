```python
import requests
from bs4 import BeautifulSoup
import csv

# Function to extract details for a section
def extract_section_text(section_name, soup):
    section_tag = soup.find('h3', text=section_name)
    if section_tag:
        section_text = ''
        next_sibling = section_tag.find_next_sibling()
        while next_sibling and next_sibling.name != 'h3':
            section_text += str(next_sibling)
            next_sibling = next_sibling.find_next_sibling()

        section_text = BeautifulSoup(section_text, 'html.parser').get_text(separator='', strip=True)
        return section_text.strip()
    else:
        print(f"Section '{section_name}' not found on the drug page.")
        return None

# Function to extract drug details
def extract_drug_details(drug_url):
    drug_response = requests.get(drug_url)
    if drug_response.status_code == 200:
        drug_soup = BeautifulSoup(drug_response.text, 'html.parser')
        # Call your existing function to extract details
        description = extract_section_text('DESCRIPTION', drug_soup)
        indications = extract_section_text('INDICATIONS', drug_soup)
        dosage_and_administration = extract_section_text('DOSAGE AND ADMINISTRATION', drug_soup)
        side_effects = extract_section_text('SIDE EFFECTS', drug_soup)
        warnings = extract_section_text('WARNINGS', drug_soup)
        precautions = extract_section_text('PRECAUTIONS', drug_soup)
        overdose = extract_section_text('OVERDOSE', drug_soup)
        contraindications = extract_section_text('CONTRAINDICATIONS', drug_soup)
        clinical_pharmacology = extract_section_text('CLINICAL PHARMACOLOGY', drug_soup)
        patient_information = extract_section_text('PATIENT INFORMATION', drug_soup)

        return {
            'DESCRIPTION': description,
            'INDICATIONS': indications,
            'DOSAGE AND ADMINISTRATION': dosage_and_administration,
            'SIDE EFFECTS': side_effects,
            'WARNINGS': warnings,
            'PRECAUTIONS': precautions,
            'OVERDOSE': overdose,
            'CONTRAINDICATIONS': contraindications,
            'CLINICAL PHARMACOLOGY': clinical_pharmacology,
            'PATIENT INFORMATION': patient_information
        }
    else:
        print(f"Failed to retrieve the drug page. Status code: {drug_response.status_code}")
        return None

# Function to extract list of drugs
def extract_list(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    data = []

    for li in soup.select('div.AZ_results ul li'):
        name = li.text.strip()
        link = li.find('a')['href']
        # Ensure the link has the correct format
        if not link.startswith('http'):
            link = f"https://www.rxlist.com{link}"
        data.append({'Name': name, 'Link': link})

    return data

# URL for the list of drugs
list_url = 'https://www.rxlist.com/drugs/alpha_b.htm'
list_data = extract_list(list_url)

# Iterate through each drug and extract details
for drug in list_data:
    drug_name = drug['Name']
    drug_link = drug['Link']
    drug_details = extract_drug_details(drug_link)

    if drug_details:
        # Add drug name to details
        drug_details['Name'] = drug_name

        # Save the details to a CSV file
        csv_file = 'drug_details.csv'
        with open(csv_file, 'a', newline='', encoding='utf-8') as csvfile:
            fieldnames = ['Name', 'DESCRIPTION', 'INDICATIONS', 'DOSAGE AND ADMINISTRATION', 'SIDE EFFECTS',
                          'WARNINGS', 'PRECAUTIONS', 'OVERDOSE', 'CONTRAINDICATIONS', 'CLINICAL PHARMACOLOGY',
                          'PATIENT INFORMATION']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

            # Write header if the file is empty
            if csvfile.tell() == 0:
                writer.writeheader()

            # Write drug details to the CSV file
            writer.writerow(drug_details)

print("Extraction and CSV creation completed.")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

# Function to extract details for a section
def extract_section_text(section_name, soup):
    section_tag = soup.find('h3', text=section_name)
    if section_tag:
        section_text = ''
        next_sibling = section_tag.find_next_sibling()
        while next_sibling and next_sibling.name != 'h3':
            section_text += str(next_sibling)
            next_sibling = next_sibling.find_next_sibling()

        section_text = BeautifulSoup(section_text, 'html.parser').get_text(separator='', strip=True)
        return section_text.strip()
    else:
        print(f"Section '{section_name}' not found on the drug page.")
        return None

# Function to extract drug details
def extract_drug_details(drug_url):
    drug_response = requests.get(drug_url)
    if drug_response.status_code == 200:
        drug_soup = BeautifulSoup(drug_response.text, 'html.parser')
        # Call your existing function to extract details
        description = extract_section_text('DESCRIPTION', drug_soup)
        indications = extract_section_text('INDICATIONS', drug_soup)
        dosage_and_administration = extract_section_text('DOSAGE AND ADMINISTRATION', drug_soup)
        side_effects = extract_section_text('SIDE EFFECTS', drug_soup)
        warnings = extract_section_text('WARNINGS', drug_soup)
        precautions = extract_section_text('PRECAUTIONS', drug_soup)
        overdose = extract_section_text('OVERDOSE', drug_soup)
        contraindications = extract_section_text('CONTRAINDICATIONS', drug_soup)
        clinical_pharmacology = extract_section_text('CLINICAL PHARMACOLOGY', drug_soup)
        patient_information = extract_section_text('PATIENT INFORMATION', drug_soup)

        return {
            'DESCRIPTION': description,
            'INDICATIONS': indications,
            'DOSAGE AND ADMINISTRATION': dosage_and_administration,
            'SIDE EFFECTS': side_effects,
            'WARNINGS': warnings,
            'PRECAUTIONS': precautions,
            'OVERDOSE': overdose,
            'CONTRAINDICATIONS': contraindications,
            'CLINICAL PHARMACOLOGY': clinical_pharmacology,
            'PATIENT INFORMATION': patient_information
        }
    else:
        print(f"Failed to retrieve the drug page. Status code: {drug_response.status_code}")
        return None

# Function to extract list of drugs
def extract_list(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    data = []

    for li in soup.select('div.AZ_results ul li'):
        name = li.text.strip()
        link = li.find('a')['href']
        # Ensure the link has the correct format
        if not link.startswith('http'):
            link = f"https://www.rxlist.com{link}"
        data.append({'Name': name, 'Link': link})

    return data

# URL for the list of drugs
list_url = 'https://www.rxlist.com/drugs/alpha_y.htm'
list_data = extract_list(list_url)

# Iterate through each drug and extract details
for drug in list_data:
    drug_name = drug['Name']
    drug_link = drug['Link']
    drug_details = extract_drug_details(drug_link)

    if drug_details:
        # Add drug name to details
        drug_details['Name'] = drug_name

        # Save the details to a CSV file
        csv_file = '/Users/b/docs/drug_details_y.csv'
        with open(csv_file, 'a', newline='', encoding='utf-8') as csvfile:
            fieldnames = ['Name', 'DESCRIPTION', 'INDICATIONS', 'DOSAGE AND ADMINISTRATION', 'SIDE EFFECTS',
                          'WARNINGS', 'PRECAUTIONS', 'OVERDOSE', 'CONTRAINDICATIONS', 'CLINICAL PHARMACOLOGY',
                          'PATIENT INFORMATION']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

            # Write header if the file is empty
            if csvfile.tell() == 0:
                writer.writeheader()

            # Write drug details to the CSV file
            writer.writerow(drug_details)

print("Extraction and CSV creation completed.")
```

```python
import pandas as pd
import os

#specify the path
path = '/Users/b/docs'

#the store for each csv file
data = []

# loop over all the files in the path
for file in os.listdir(path):
    if file.endswith('.csv'):
       file_path = os.path.join(path, file)
       try:  
            df = pd.read_csv(file_path)
            data.append(df)
       except Exception as e:
            print(f"Error reading file {file}: {e}")
       
#concatenate all the dataframes
combined_df = pd.concat(data, ignore_index=True)

# convert to json
output_json = 'famasi_drugs_db.json'
combined_df.to_json(output_json, orient='records', indent=4, lines=True)
print ("JSON file created successfully")
```
